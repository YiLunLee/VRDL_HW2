```python
import h5py
import numpy as np
import os
import cv2

with h5py.File('./vrdl_data/train/digitStruct.mat', 'r') as f:
    f.keys()
    data = f.get('digitStruct')
    # name
    data = np.array(f['digitStruct']['name']) # For converting to a NumPy array
    bbox = np.array(f['digitStruct']['bbox']) # For converting to a NumPy array
    image_root = './vrdl_data/images/train/'
    for i in range(data.shape[0]):
        if i % 100 == 0:
            print("finished ", i)
        i = 12667
        # image name
        x = data[i][0]
        name = ''.join(chr(k) for k in f[x][:])
        image_path = os.path.join(image_root, name)
        img = cv2.imread(image_path)
        image_h, image_w, _ = img.shape
        # image bbox
        bbox = f['digitStruct']['bbox'][i][0]
        label_path = os.path.join(image_root.replace('images','labels'), name.replace('.png', '.txt'))
        
        with open(label_path, 'w') as fout:
            for j in range(f[bbox]['label'].shape[0]):
                if f[bbox]['label'].shape[0] == 1:
                    height = f[bbox]['height'][0][0]
                    label = f[bbox]['label'][0][0]
                    left = f[bbox]['left'][0][0]
                    top = f[bbox]['top'][0][0]
                    width = f[bbox]['width'][0][0]
                else:
                    height = f[f[bbox]['height'][j][0]][0][0]
                    label = f[f[bbox]['label'][j][0]][0][0]
                    left = f[f[bbox]['left'][j][0]][0][0]
                    top = f[f[bbox]['top'][j][0]][0][0]
                    width = f[f[bbox]['width'][j][0]][0][0]
                center_x = (left + width/2) / image_w
                center_y = (top + height/2) / image_h
                width = width / image_w
                height = height / image_h
                label = int(label) if label < 10 else 0
                if width > 1.0 or width > 1.0 or center_y > 1.0 or center_x > 1.0:
                    print('Fuck you {} {} {} {} {}\n'.format(label, center_x, center_y, width, height))
                    break
                labels = '{} {} {} {} {}\n'.format(label, center_x, center_y, width, height)
                fout.write(labels)
            fout.close()
        break
```

```python
import os
import random
image_root = './vrdl_data/images/train'
train_path = './vrdl_data/images/train_set'
val_path = './vrdl_data/images/val'
images = os.listdir('./vrdl_data/images/train')
# print(images)
val_images = random.sample(images,3402)
print(val_images)
for img in images:
    src_img_path = os.path.join(image_root,img)
    src_label_path = os.path.join(image_root.replace('images','labels'),img.replace('.png','.txt'))
    if img in val_images:
        target_img_path = src_img_path.replace('train', 'val')
        target_label_path = src_label_path.replace('train', 'val')
    else:
        target_img_path = src_img_path.replace('train', 'train_set')
        target_label_path = src_label_path.replace('train', 'train_set')
    os.system('cp {} {}'.format(src_img_path, target_img_path))
    os.system('cp {} {}'.format(src_label_path, target_label_path))
    
```
